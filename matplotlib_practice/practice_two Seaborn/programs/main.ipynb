*Working with **Seaborn***

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../files/train.csv')
```

```python
plt.figure(
    figsize=(10, 10),
    dpi=100,
)


plt.title("Survival Distribution")
sns.lineplot(
    data=data,
    x="Age",
    y="Survived",
    
)

plt.show()
```

*Fare vs Pclass*

```python
plt.figure(
    figsize=(6, 6),
    dpi=100,
)
plt.title("Fare vs Pclass")

sns.lineplot(
    data=data,
    x="Pclass",
    y="Fare",
)

plt.xticks([1,2,3])
plt.show()
```

***Dis Plot***

```python
sns.displot(
    data["Age"],
    kde=True,
    alpha= 0.5,
)
```

***Scatter Plot***

```python
plt.figure(figsize=(7,7), dpi = 100)
sns.scatterplot(
    data=data,
    x="Pclass",
    y="Fare",
    alpha=0.6,
)

plt.xticks([1,2,3])

plt.show()
```

```python
plt.figure(figsize=(7,7), dpi = 100)
plt.title('Fare vs Pclass')
sns.scatterplot(
    data=data,
    x="Pclass",
    y="Fare",
    alpha=0.6,
    hue= 'Survived',
    palette='inferno'
)

plt.xticks([1,2,3])

plt.show()
```

***Bar Plot***

*A Bar Plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using errors bars.*

```python
sns.barplot(
    x="Pclass",
    y="Fare",
    data=data,
    hue="Sex",
    palette='inferno'
)
```

```python
# orient
# to use horizontal orientation, x must be numeric_value and y can be anything

data_temp = data.sample(n=15)

sns.barplot(
    x="Pclass",
    y="Fare",
    data=data_temp,
    orient="h",
)

plt.xticks([1, 2, 3])

plt.show()
```

```python
# estimator

sns.barplot(
    x="Pclass",
    y="Fare",
    data=data,
    hue="Sex",
    estimator=np.max,
)

plt.xticks([1, 2, 3])

plt.show()
```

```python
# confidence interval

# err_kws is a dictionary where you can specify keyword arguments for the error bar styling (e.g., linewidth, alpha, etc.).

sns.barplot(
    x="Pclass",
    y="Fare",
    data=data,
    errorbar=("ci", 75),
    err_kws={
        "color": "black",
        "linewidth": 5,
    },
)
```

```python
# saturation
sns.barplot(
    x="Pclass",
    y="Fare",
    data=data,
    saturation= 0.5,
    errorbar=("ci", 75),
    err_kws={
        "color": "black",
        "linewidth": 5,
    },
)
```

***HeatMap***

*The **annot** parameter determines whether to display the numeric values on the heatmap. By default, this is set to False, meaning that no numeric annotations are shown. If you set annot=True, it will display the data values inside the heatmap cells.*

*the **cmap** parameter controls the colormap used to color the cells of the heatmap. A colormap defines the mapping of data values (e.g., numbers) to colors. The cmap parameter allows you to specify how the data values are represented visually with different colors.*

```python
temp_data = data.pivot_table(
    index="Pclass",
    columns="Sex",
    values="Survived",
    aggfunc="mean",
)

plt.figure(
    figsize=(6, 6),
    dpi=150,
)

plt.title("Heatmap")
sns.heatmap(
    temp_data,
    annot=True,
    fmt=".2f",
    linecolor='w',
    linewidths=2,
    cmap= 'Blues',
)

plt.show()
```

```python
sns.heatmap(
    temp_data.corr(),
    annot=True,
    fmt=".2f",
    cmap="Blues",
)
```
