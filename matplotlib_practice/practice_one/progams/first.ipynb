# importing all the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

*creating a basic graph* 

```python
x = [1, 2, 3, 4]
y = [10, 4, 9, 11]

# resizing graph
plt.figure(figsize=(5, 5), dpi=200)

# line graph one
plt.plot(
    x,
    y,
    marker="o",
    markersize="5",
    markeredgecolor="white",
    label="Random",
    color="black",
    linewidth=2,
    linestyle=":",
)

# line graph 2
a = np.arange(0, 5, 2)
plt.plot(
    a,
    a**2,
    color="red",
    marker="o",
    markeredgecolor="white",
    linestyle="-",
    label="a sq",
)

# line graph 3
b = np.arange(0, 5, 1)
plt.plot(
    b[:4],
    b[:4] ** 2,
    marker="o",
    markeredgecolor="white",
    color="blue",
    linestyle="--",
    label="b sq till index 3",
)
plt.plot(
    b[3:],
    b[3:] ** 2,
    marker="o",
    markeredgecolor="white",
    linestyle="-",
    color="blue",
    label="b sq from index 3",
)

# X and Y labels
plt.xlabel("Numbers of x")
plt.ylabel("Numbers of y")

# title of the graph
plt.title("Line Graph", fontdict={"fontsize": 20})

# X and Y ticks (scale of graph)
plt.xticks([1, 2, 3, 4, 5])
# plt.yticks([10,7,0,9,11])

# saving the graph
plt.savefig(
    "../files/linegraph.png",
    dpi=300,
)

# aadding a legend
plt.legend()  # legend takes the value to show from the label we give to describe the pattern or how the data is kept
plt.show()
```

# BAR CHART

```python
xlabels = ['A', 'B', 'C']
ylabels = [2, 10, 4]

# giving the size of bar chart
plt.figure(figsize=(6, 6))

bars = plt.bar(xlabels, ylabels, label = 'ABC')

# setting hatching lines
# bars[0].set_hatch('/')
# bars[1].set_hatch("+")
# bars[2].set_hatch("-")

patterns = ['/', '+', '-']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.legend()
plt.savefig('../files/barchcart.png', dpi = 300)
plt.show()
```
